|Variable|Definition|Key|Notes|
|:----:|:----:|:----:|:----:|
|survival|Survival|0 = No, 1 = Yes|null|
|pclass|Ticket class|1 = 1st, 2 = 2nd, 3 = 3rd|A proxy for socio-economic status (SES). 1st = Upper. 2nd = Middle. 3rd = Lower|
|sex|sex|male, female||
|Age|Age in years||Age is fractional if less than 1. If the age is estimated, is it in the form of xx.5|
|sibsp|# of siblings / spouses aboard the Titanic||Sibling = brother, sister, stepbrother, stepsister. Spouse = husband, wife (mistresses and fiancés were ignored|
|parch|# of parents / children aboard the Titanic||Parent = mother, father. Child = daughter, son, stepdaughter, stepson. Some children travelled only with a nanny, therefore parch=0 for them.|
|fare|Passenger fare|||
|cabin|Cabin number|||
|embarked|Port of Embarkation	C = Cherbourg, Q = Queenstown, S = Southampton|||


```python
# Bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
import seaborn as sns
```

## Esta análise foi dividida em alguns blocos que são descritos a seguir

### O primeiro bloco é relativo a importação e tratamento dos dados

```python
# Importando as bases originais
ts = pd.read_csv('./titanic1_submission.csv')
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
```

```python
ts.head(3)
```

```python
train.head(3)
```

```python
train.info()
```

```python
test.head(3)
```

```python
test.info()
```

```python
# Como para esta etapa do projeto não há necessidade de divisão entre as bases de treino e teste, as mesmas serão unificadas.
#1. incluindo a variárel "survived" a base de teste:
test2=pd.merge(test, ts, on="PassengerId", how="left")
```

```python
test2.head()
```

```python
test2.info()
```

```python
#juntando as duas bases, teste e treino num único dataframe
df=test2.append(train, ignore_index = True)
```

```python
df.head(3)
```

```python
# criando uma nova coluna family, cujo objetivo é marcar se havia no navio algum familiar, ou seja, junção entre sibsp e parch
df['has_family_onboard'] = np.where((df['SibSp']>0) | (df['Parch']>0), 'Yes', 'No')
df.head()
```

```python
#Criando uma nova coluna, com base no nome, para identififcar se o passageiro era casado ou solteiro.
married_condition = (df['Name'].str.contains('Mr.', regex=False)) | (df['Name'].str.contains('Mrs.', regex=False))
df['married'] = np.where(married_condition, 'Yes', 'No')
df.head()
```

### O segundo bloco contém as análises univariadas.

```python
# Distribuição de idade
df_age = df[~df['Age'].isnull()]
df_age.plot(y='Age', kind='box', title= 'Dist. de Idade')
age_hist = df_age.plot(y='Age', kind='hist', bins=30);
age_hist.set_xlabel('Idade dos passageiros', fontsize=10);
```

é identificado alguns outliers, que são pessoas com idade acima de 60 e poucos anos. Não iremos realizar nenhum tratamento nesses casos, pois são idades corretas, porém somente com pouca representatividade na base estudada.

Com os gráficos é possível identificar uma concentração de passageiros com idade entre 20 e 40 anos, além da presença de crianças abaixo de 10 anos.

```python
# proporção de homens e mulheres
df['Sex'].value_counts().plot(kind='pie',autopct='%1.0f%%', legend=True, title = ' Freq. Rel. de Sexo');
```

Havia mais homens do que mulheres no navio. Perguntas que serão respondidas mais adiante: será que essa distribuição se mantém em todas as classes do navio?

```python
df1=df.loc[df['Pclass'] == 1]
df2=df.loc[df['Pclass'] == 2]
df3=df.loc[df['Pclass'] == 3]
df1['Sex'].value_counts().plot(kind='pie',autopct='%1.0f%%', legend=True, title = ' Freq. Rel. de Sexo - Primeira classe')


```

```python
df2['Sex'].value_counts().plot(kind='pie',autopct='%1.0f%%', legend=True, title = ' Freq. Rel. de Sexo - Segunda classe')
```

```python
df3['Sex'].value_counts().plot(kind='pie',autopct='%1.0f%%', legend=True, title = ' Freq. Rel. de Sexo - Terceira classe');
```

Notamos que conforme vai diminuindo a classe, a proporção de homens aumentam.

```python
# Freq de embarques em cada porto
df['Embarked'].value_counts().plot(kind='bar',title = 'Freq. Embarque por porto');
```

Segundo a https://pt.wikipedia.org/wiki/RMS_Titanic o porto de saída foi o Southampton , o que aqui identiifcamos ser o porto com maior quantidade de passageiros.

```python
# Freq dos tipos de classe
# df['Pclass'].value_counts().plot(kind='bar',title = 'Freq. dos tipos de Classe');
# plt.bar()

# Mudar Pclass para string, para tirar os decimais e o 0.5
bar_data = df['Pclass'].apply(str).value_counts()
plt.title('Freq. dos tipos de Classe')
plt.barh(bar_data.index.values, bar_data.values)
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%1.0f%%', legend=True, title = 'Freq. Rel. dos tipos de Classe');
```

A maioria dos passageiros eram da terceira classe.

```python
# proporção de sobreviventes
df['Survived'].value_counts().plot(kind='pie',autopct='%1.0f%%', legend=True, title = ' Freq. de sobreviventes');
```

A maioria dos passageiros não sobreviveram

# Bloco para identificar relações entre sobreviventes e não sobreviventes

```python
# Idade por sobreviventes e não sobreviventes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
plt.figure(figsize=(15,2))
ax1.hist(bins=10, x=df[df['Survived'] == 1]['Age'], alpha=0.5, color='blue')
#ax1.set_ylim([0, 600])
ax2.hist(bins=10, x=df[df['Survived'] == 0]['Age'], alpha=0.5, color='green')
#ax2.set_ylim([0, 600])
```

```python
# Tipo de classe por sobreviventes e não sobreviventes
df[df['Survived'] == 0]['Pclass'].value_counts().plot(kind='bar', legend=True, title = ' Freq. de Não Sobreviventes por Classe')
```

```python
df[df['Survived'] == 0]['Pclass'].value_counts().plot(kind='pie',autopct='%1.0f%%', legend=True, title = ' Freq. de Não Sobreviventes por Classe');
```

```python
df[df['Survived'] == 1]['Pclass'].value_counts().plot(kind='bar', legend=True, title = ' Freq. de Sobreviventes por Classe');
```

```python
df[df['Survived'] == 1]['Pclass'].value_counts().plot(kind='pie',autopct='%1.0f%%', legend=True, title = ' Freq. de Sobreviventes por Classe');
```

Mais da metade dos sobreviventes eram das classes 1 e 2, já dentre os não sobreviventes, a grande maioria, eram da terceira classe

```python
# Sexo por sobreviventes e não sobreviventes
df[df['Survived'] == 0]['Sex'].value_counts().plot(kind='pie',autopct='%1.0f%%', legend=True, title = ' Freq. de Não Sobreviventes por Sexo');
```

```python
df[df['Survived'] == 1]['Sex'].value_counts().plot(kind='pie',autopct='%1.0f%%', legend=True, title = ' Freq. de Sobreviventes por Sexo');
```

Notamos uma "prevalência" de mulheres dentre as sobreviventes. Abaixo veremos se a composição sexo+classe, teve alguma diferença na proporção dentre os sobreviventes e não sobreviventes.

por este gráfico, realmente o Jack tinha menos "chance" de sobreviver do que a Rose

```python
# sobreviventes por posse de "parentes" 
df[df['Survived'] == 1]['has_family_onboard'].value_counts().plot(kind='pie',autopct='%1.0f%%', legend=True, title = ' Freq. de Sobreviventes por posse de parentes');
```

```python
df[df['Survived'] == 0]['has_family_onboard'].value_counts().plot(kind='pie',autopct='%1.0f%%', legend=True, title = ' Freq. de Não Sobreviventes por posse de parentes');
```

 ter família a bordo não foi um fator que favoreceu a sobrevivência. o fato de ter dado maior proporção de sem familias dentre os não sobreviventes, pode ser devido a classe.
